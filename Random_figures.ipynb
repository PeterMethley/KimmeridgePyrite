```python
from plotly_default import go
```

```python
cb_test = go.Figure()
cb_test.add_trace(go.Scatter(x=[1,2], y=[1,2], marker_color=[0.94,22.21], marker_colorscale='viridis', marker_colorbar_title='Height / m', marker_colorbar_ticks='outside'))
cb_test.update_layout(width=800, height=700)
cb_test.show()
```
